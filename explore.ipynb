```python
from APIs.cash_flow import *
from APIs.income_statement import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
cash_raw = get_cash_flow('AAPL')
```

```python
cash_df = pd.DataFrame(cash_raw)
# convert fiscal date ending to datetime
cash_df['fiscalDateEnding'] = pd.to_datetime(cash_df['fiscalDateEnding'])
# convert netIncome to numeric
cash_df['netIncome'] = pd.to_numeric(cash_df['netIncome'], errors='coerce')
cash_df['capitalExpenditures'] = pd.to_numeric(cash_df['capitalExpenditures'], errors='coerce')
cash_df['operatingCashflow'] = pd.to_numeric(cash_df['operatingCashflow'], errors='coerce')
cash_df.head()
```

```python
cash_df.plot(x='fiscalDateEnding', y=['netIncome', 'capitalExpenditures', 'operatingCashflow'], kind='line', figsize=(10, 6))
```

```python
income_raw = get_income_statement('AAPL')
income_df = pd.DataFrame(income_raw)
income_df['fiscalDateEnding'] = pd.to_datetime(income_df['fiscalDateEnding'])
income_df['grossProfit'] = pd.to_numeric(income_df['grossProfit'], errors='coerce')
income_df['totalRevenue'] = pd.to_numeric(income_df['totalRevenue'], errors='coerce')
income_df.head()
```

```python
income_df.plot(x='fiscalDateEnding', y=['grossProfit', 'totalRevenue'], kind='line', figsize=(10, 6))
```
